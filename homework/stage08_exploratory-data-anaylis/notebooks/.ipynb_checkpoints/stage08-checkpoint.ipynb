```python
#importing this cell from homewrok starter
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from scipy.stats import skew, kurtosis
sns.set(context='talk', style='whitegrid')
np.random.seed(8)
pd.set_option('display.max_columns', 100)

n = 160
df = pd.DataFrame({
    'date': pd.date_range('2021-02-01', periods=n, freq='D'),
    'region': np.random.choice(['North','South','East','West'], size=n),
    'age': np.random.normal(40, 8, size=n).clip(22, 70).round(1),
    'income': np.random.lognormal(mean=10.6, sigma=0.3, size=n).round(2),
    'transactions': np.random.poisson(lam=3, size=n),
})
base = df['income'] * 0.0015 + df['transactions']*18 + np.random.normal(0, 40, size=n)
df['spend'] = np.maximum(0, base).round(2)

# inject a bit of missingness and outliers
df.loc[np.random.choice(df.index, 5, replace=False), 'income'] = np.nan
df.loc[np.random.choice(df.index, 3, replace=False), 'spend'] = np.nan
df.loc[np.random.choice(df.index, 2, replace=False), 'transactions'] = df['transactions'].max()+12
df.head()
```

```python
#Distributions
# Histogram for income 
plt.figure(figsize=(6,4))
df['income'].hist(bins=30)
plt.title("Histogram: income")
plt.show()

# Boxplot for transactions
plt.figure(figsize=(6,4))
plt.boxplot(df['transactions'])
plt.title("Boxplot: transactions")
plt.show()

```

```python
# Histogram for age
plt.figure(figsize=(6,4))
df['age'].hist(bins=30)
plt.title("Histogram: Age")
plt.show()

```

```python
#Bivariate realtionship
plt.figure(figsize=(6,4))
plt.scatter(df['age'], df['spend'])
plt.title("Scatter: Age vs How much they spend weekly on average")
plt.xlabel("age")
plt.ylabel("spend in dollars")
plt.show()
```

```python
plt.figure
plt.scatter(df['transactions'], df['income'])
plt.title("Scatter: Transactions vs income")
plt.xlabel("transactions")
plt.ylabel("income in dollars")
plt.show()
```

```python
#Heatmap
import seaborn as sns

plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Findings:

- Age vs how much people spend weekly also a few outliers but is mostly normal.
- Strong positive correlation (0.5) between transactions and spend.

### Risks:
